```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_data_comparison4(
        plot_title: str,
        data1: list[tuple[int, float]],
        data1_label: str,
        data2: list[tuple[int, float]],
        data2_label: str,
        data3: list[tuple[int, float]],
        data3_label: str,
        data4: list[tuple[int, float]],
        data4_label: str,
        x_label: str,
        y_label: str,
        log_scale: bool = True):
    x = [x for x, _ in data1]
    y1 = [y for _, y in data1]
    y2 = [y for _, y in data2]
    y3 = [y for _, y in data3]
    y4 = [y for _, y in data4]

    plt.plot(x, y1, label=data1_label, color='tab:blue', linestyle="", marker="o")
    plt.plot(x, y2, label=data2_label, color='tab:orange', linestyle="", marker="o")
    plt.plot(x, y3, label=data3_label, color='black', linestyle="", marker="o")
    plt.plot(x, y4, label=data4_label, color='tab:green', linestyle="", marker="o")

    # Set to log scale
    plt.yscale('log') if log_scale else None
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(plot_title)
    plt.legend()

    plt.show()

def plot_data_comparison3(
        plot_title: str, 
        data1: list[tuple[int, float]], 
        data1_label: str,
        data2: list[tuple[int, float]], 
        data2_label: str,
        data3: list[tuple[int, float]],
        data3_label: str,
        x_label: str, 
        y_label: str,
        log_scale: bool = True):
    x = [x for x, _ in data1]
    y1 = [y for _, y in data1]
    y2 = [y for _, y in data2]
    y3 = [y for _, y in data3]

    # plotting points for the different data points
    plt.plot(x, y1, label=data1_label, color='tab:blue', linestyle="", marker="o")
    plt.plot(x, y2, label=data2_label, color='tab:orange' , linestyle="", marker="o")
    plt.plot(x, y3, label=data3_label, color='black', linestyle="", marker="o")

    # Set to log scale
    plt.yscale('log') if log_scale else None
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(plot_title)
    plt.legend()
    plt.show()

def plot_data_comparison2(
        plot_title: str, 
        data1: list[tuple[int, float]], 
        data1_label: str,
        data2: list[tuple[int, float]], 
        data2_label: str,
        x_label: str, 
        y_label: str,
        log_scale: bool = True):
    x = [x for x, _ in data1]
    y1 = [y for _, y in data1]
    y2 = [y for _, y in data2]

    plt.plot(x, y1, label=data1_label, linestyle="", marker="o")
    plt.plot(x, y2, label=data2_label, linestyle="", marker="o")
    # Set to log scale
    plt.yscale('log') if log_scale else None
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(plot_title)
    plt.legend()
    plt.show()

def plot_data_comparison2_with_error_bars(
        plot_title: str,
        data1: list[tuple[int, tuple[float, float]]],
        data1_label: str,
        data2: list[tuple[int, tuple[float, float]]],
        data2_label: str,
        x_label: str,
        y_label: str,
        log_scale: bool = True):
    
    x = [x for x, _ in data1]
    y1 = [y for _, (y, _) in data1]
    y1_err = [err for _, (_, err) in data1]
    y2 = [y for _, (y, _) in data2]
    y2_err = [err for _, (_, err) in data2]

    plt.errorbar(x, y1, yerr=y1_err, label=data1_label, linestyle="", marker="o", capsize=5)
    plt.errorbar(x, y2, yerr=y2_err, label=data2_label, linestyle="", marker="^", capsize=5)
    # Set to log scale
    plt.yscale('log') if log_scale else None
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(plot_title)
    plt.legend()
    plt.show()
    
  

def plot_data_with_regression(plot_title, data: list[tuple[int, float]], data_label: str, x_label: str, y_label: str, degree: int):
    x = [x for x, _ in data]
    y = [y for _, y in data]
    plt.plot(x, y, label=data_label, linestyle="", marker="o")
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    z = np.polyfit(x, y, degree)
    p = np.poly1d(z)
    plt.plot(x, p(x), "r--")    
    plt.title(plot_title + ", with regression degree " + str(degree))
    plt.legend()
    plt.show()
```
