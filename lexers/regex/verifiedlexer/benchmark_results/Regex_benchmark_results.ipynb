```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_data_comparison3(
        plot_title: str, 
        data1: list[tuple[int, float]], 
        data1_label: str,
        data2: list[tuple[int, float]], 
        data2_label: str,
        data3: list[tuple[int, float]],
        data3_label: str,
        x_label: str, 
        y_label: str):
    x = [x for x, _ in data1]
    y1 = [y for _, y in data1]
    y2 = [y for _, y in data2]
    y3 = [y for _, y in data3]

    plt.plot(x, y1, label=data1_label, color='tab:blue')
    plt.plot(x, y2, label=data2_label, color='tab:orange')
    plt.plot(x, y3, label=data3_label, color='black')

    # Set to log scale
    plt.yscale('log')
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(plot_title)
    plt.legend()
    plt.show()

def plot_data_comparison2(
        plot_title: str, 
        data1: list[tuple[int, float]], 
        data1_label: str,
        data2: list[tuple[int, float]], 
        data2_label: str,
        x_label: str, 
        y_label: str):
    x = [x for x, _ in data1]
    y1 = [y for _, y in data1]
    y2 = [y for _, y in data2]

    plt.plot(x, y1, label=data1_label)
    plt.plot(x, y2, label=data2_label)
    # Set to log scale
    plt.yscale('log')
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(plot_title)
    plt.legend()
    plt.show()
  

def plot_data(plot_title, data: list[tuple[int, float]], data_label: str, x_label: str, y_label: str):
    x = [x for x, _ in data]
    y = [y for _, y in data]
    plt.plot(x, y, label=data_label)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(plot_title)
    plt.legend()
    plt.show()
```

```python
plot_data_comparison3(
  plot_title = "Matching regex (a|b)* against accepting strings", 
  data1 = regex_data,
  data1_label = "Regex",
  data2 = zipper_data,
  data2_label = "Zipper",
  data3 = regex_mem_data,
  data3_label = "Regex with memoization",
  x_label = "String length",
  y_label = "Time (us)")
```
